## Thực hiện gộp các file data crawl được

```python
import os
import pandas as pd

def merge_csv_files(input_folder_1, input_folder_2, output_folder_1, output_folder_2, output_file_name_1, output_file_name_2):
    if not os.path.exists(input_folder_1):
        print(f"Thư mục {input_folder_1} không tồn tại, đang tạo thư mục...")
        os.makedirs(input_folder_1)

    if not os.path.exists(input_folder_2):
        print(f"Thư mục {input_folder_2} không tồn tại, đang tạo thư mục...")
        os.makedirs(input_folder_2)

    files_1 = [file for file in os.listdir(input_folder_1) if file.endswith('.csv')]
    files_2 = [file for file in os.listdir(input_folder_2) if file.endswith('.csv')]

    merge_files(files_1, input_folder_1, output_folder_1, output_file_name_1)
    merge_files(files_2, input_folder_2, output_folder_2, output_file_name_2)

def merge_files(files, input_folder, output_folder, output_file_name):
    if len(files) == 0:
        print(f"Không có file CSV nào trong thư mục {input_folder}.")
        return

    merged_df = pd.DataFrame()

    for i, file in enumerate(files):
        file_path = os.path.join(input_folder, file)
        df = pd.read_csv(file_path)

        if not df.empty:
            if i == 0:
                merged_df = df
            else:
                merged_df = pd.concat([merged_df, df], ignore_index=True)

    if not merged_df.empty:
        output_path = os.path.join(output_folder, output_file_name)
        merged_df.to_csv(output_path, index=False)
        print(f"Gộp thành công! File đã gộp được lưu tại {output_path}.")
    else:
        print("Không có dữ liệu để gộp.")

# Thư mục chứa các file CSV cần gộp từ nguồn dữ liệu 1
input_folder_1 = 'D:\\LNTP ở HUST\\Học tập\Năm 3\\20232\\Đồ án 1\\Source\\Data-Crawling-Of-Amazon-main\\DataSource\\Part 1'
# Thư mục chứa các file CSV cần gộp từ nguồn dữ liệu 2
input_folder_2 = 'D:\\LNTP ở HUST\\Học tập\Năm 3\\20232\\Đồ án 1\\Source\\Data-Crawling-Of-Amazon-main\\DataSource\\Part 2'
# Thư mục chứa file output sau khi gộp từ nguồn dữ liệu 1
output_folder_1 = 'D:\\LNTP ở HUST\\Học tập\Năm 3\\20232\\Đồ án 1\\Source\\Data-Crawling-Of-Amazon-main\\DataSource\\Part 1\\Part1AfterMerge'
# Thư mục chứa file output sau khi gộp từ nguồn dữ liệu 2
output_folder_2 = 'D:\\LNTP ở HUST\\Học tập\Năm 3\\20232\\Đồ án 1\\Source\\Data-Crawling-Of-Amazon-main\\DataSource\\Part 2\\Part2AfterMerge'
# Tên của file output sau khi gộp từ nguồn dữ liệu 1
output_file_name_1 = 'Data_Part1.csv'
# Tên của file output sau khi gộp từ nguồn dữ liệu 2
output_file_name_2 = 'Data_Part2.csv'

# Gọi hàm merge_csv_files
merge_csv_files(input_folder_1, input_folder_2, output_folder_1, output_folder_2, output_file_name_1, output_file_name_2)
```

```python
def convert_weight_to_ounces(weight_str):
    parts = weight_str.split()
    if len(parts) == 2:
        weight = float(parts[0])
        unit = parts[1]
        if unit == 'pounds':
            return weight * 16  # 1 pound = 16 ounces
        elif unit == 'ounces':
            return weight
    # Trả về giá trị ban đầu nếu định dạng không như mong đợi
    return weight_str

def convert_column_to_ounces(df, column_name):
    df[column_name] = df[column_name].apply(convert_weight_to_ounces)
    df[column_name] = df[column_name].astype(str) + ' ounces'
    return df

# Gọi hàm convert_column_to_ounces với DataFrame và tên cột "Item Weight"
df_part2 = convert_column_to_ounces(df_part2, "Item Weight")
```

#### Tách các Main ranking và Sub ranking

```python
def process_ranking(data):
    # Tìm vị trí của dấu "#" đầu tiên trong chuỗi
    first_hash_index = data.find('#')
    
    if first_hash_index != -1:
        # Tìm vị trí của dấu "#" tiếp theo trong chuỗi
        second_hash_index = data.find('#', first_hash_index + 1)
        if second_hash_index != -1:
            # Tìm vị trí của dấu "#" thứ ba trong chuỗi
            third_hash_index = data.find('#', second_hash_index + 1)
            
            if third_hash_index != -1:
                # Nếu có dấu "#" thứ ba, chỉ lấy dữ liệu đến dấu "#" thứ hai
                main_ranking = data[:second_hash_index].strip()
                sub_ranking = data[second_hash_index:third_hash_index].strip()
            else:
                # Nếu chỉ có hai dấu "#", tách dữ liệu thành phần 'Main ranking' và 'Sub ranking'
                main_ranking = data[:second_hash_index].strip()
                sub_ranking = data[second_hash_index:].strip()
        else:
            # Nếu chỉ có một dấu "#", đưa toàn bộ dữ liệu vào 'Main ranking' và 'Sub ranking' rỗng
            main_ranking = data.strip()
            sub_ranking = ''
    else:
        # Nếu không tìm thấy dấu "#", giữ toàn bộ chuỗi trong 'Main ranking' và gán 'Sub ranking' là rỗng
        main_ranking = data.strip()
        sub_ranking = ''
    
    return main_ranking, sub_ranking
```

```python
# Áp dụng hàm process_ranking vào cột 'Best Sellers Rank' trong DataFrame df_part2
df_part2['Main Ranking'], df_part2['Sub Ranking'] = zip(*df_part2['Best Sellers Rank'].apply(process_ranking))

# Xóa cột 'Best Sellers Rank' khỏi DataFrame df_part2
del df_part2['Best Sellers Rank']
```

```python
# Hàm để chuyển đổi chuỗi ngày tháng sang định dạng ngày tháng
def convert_to_date(date_string):
    try:
        # Chuyển đổi chuỗi thành định dạng ngày tháng
        date_object = datetime.strptime(date_string, '%B %d, %Y')
        return date_object
    except ValueError:
        # Trong trường hợp có lỗi khi chuyển đổi, trả về None
        return None

```
